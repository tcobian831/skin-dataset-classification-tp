Procedimiento:

1. Se toman los hiperparámetros elegidos por validación cruzada.
2. Se reentrenan los modelos con todo `trainval`.
3. Se evalúa una única vez sobre `test`.

Modelos evaluados:

- mejor MLP;
- mejor CNN entrenada desde cero;
- mejor modelo con transfer learning.

El test no se usa para elegir hiperparámetros.
